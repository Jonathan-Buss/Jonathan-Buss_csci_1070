23. (3 pts) Write a function that takes in any word and counts the number of a particular letter in it. Give the function a default value if no word is passed in.

```python
def count_letter(word=None, letter="s"):
    if word is None:
        print("oops!")
        return 0
    return word.count(letter)

count_letter("criss-cross applesauce", "c")


```
