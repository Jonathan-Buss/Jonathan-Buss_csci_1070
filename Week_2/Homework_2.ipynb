4. (3 pts) Create a function that prints a triangle of asterisks as below. 
The function accepts n as an integer and makes the longest leg of the triangle that length, decrementing by one until you close the triangle. The function also accepts a direction for the triangle as a variable. If you pass in "base-up" it returns the first, but "base-down" returns the second.
For example, for a triangle where n = 5, the following would be produced:

```python
#Base-up triangle
# * * * * *
# * * * *
# * * *
# * *
# *

#Base-down triangle
# *
# * *
# * * *
# * * * *
# * * * * *

def triangles(n, direction):
    #your code here
    if direction == "base-up":
        for i in range(n, 0, -1):
            print("* " * i)
    elif direction == "base-down":
        for i in range(1, n + 1):
            print("* " * i)

triangles(5, "base-up")

```

```python
# your code here
def letter_to_value(Letter):
    if not Letter.isalpha() or len(Letter) != 1: 
        print("ERROR")
        return "ERROR"

    letter = Letter.upper()

    goal = ord(letter) - ord("A") + 1

    for a in range(8,-1,-1):
        for b in range(8,-1,-1):
            for c in range(8,-1,-1):
                for d in range(8,-1,-1):
                    if a + b + c + d == goal:
                        result = f"a = {a}, b = {b}, c = {c}, d = {d} for letter {letter}"    
                        print(result)
                        return result
letter_to_value("J")
                        
```

What is a static method? (1 pt for your answer here) 

```python

# A static method is a function that operates inside a class but does not change the class or the instances inside it. 
# It does not use "self"or "cls" 

class actress():
    def __init__(self, name = 'Meryl Streep'):
        self.name = name
        
    def act(self, line):
        print(f'{self.name}: {line}')
        
    # your static method goes here.
    @staticmethod
    def iconic_line():
        return "*Willhelm Scream*"
        
name = actress()
name.act('This is my line...')
print(actress.iconic_line())
```

9. (3 pts) Create a class called `Dice` that accepts a single argument called `number_of_dice`. 

Create an instance method called `roll`. 

`roll` should randomly pick a number for each die when called and print the results as a list. 

Create an instance of `Dice`. Then call `roll()` on your instance for each die. 

Hint: Inside your class create an empty list called `dice`. Use [randint](https://docs.python.org/3/library/random.html#random.randrange) function to randomly generate a function between 1 and 6

```python
# Helpful starter code
from random import randint
randint(1, 6) 

class Dice():
    def __init__(self, number_of_dice):
        self.number_of_dice = number_of_dice
    def roll(self):
        results = []
        for _ in range(self.number_of_dice):
            results.append(randint(1, 6))
            return results
play = Dice(17)
event = play.roll()
print(event)
```
